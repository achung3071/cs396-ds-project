## Chi Square Test

```python
g['INFLATION_ADJ_SALARY'].mean()
```

```python
underless, overless = g[g['PTS'] <= 6.5].groupby(g['INFLATION_ADJ_SALARY'] < 4e+06).size()
undergreater, overgreater = g[g['PTS'] >= 6.5].groupby(g['INFLATION_ADJ_SALARY'] > 4e+06).size()
table = [[underless, overless],[undergreater, overgreater]]
print(table[0])
print(table[1])
```

```python
scipy.stats.chi2_contingency(table)
```
