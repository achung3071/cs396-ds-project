## Preprocessing for Model Training

```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("../data/clean/stats_salaries.csv")
df
```

It doesn't look like we need the `PLAYER_NAME` or `PLAYER_ID` for training, so we can drop that.

```python
df = df.drop(columns=['PLAYER_NAME', 'PLAYER_ID'])
df
```

We also convert minutes played to just seconds:

```python
df['SECS_PLAYED'] = df['MIN'].apply(lambda x: int(x.split(':')[0]) * 60 + int(x.split(':')[1]))
df = df.drop(columns=['MIN'])
df
```

We also convert `START_POSITION` to a dummy variable that shows the starting position of the player. If the player did not start, he will have a '0' value in all of the dummy variable columns.

```python
positions = pd.get_dummies(df['START_POSITION'])
df = pd.concat([df, positions], axis=1).drop(columns=['START_POSITION'])
df
```

We also convert the team abbreviation to dummy variables. We expect this could have an effect since different teams have different salary caps.

```python
teams = pd.get_dummies(df['TEAM_ABBREVIATION'])
df = pd.concat([df, teams], axis=1).drop(columns=['TEAM_ABBREVIATION', 'TEAM_NAME'])
df
```

For `SEASON_START` data, we can either handle this as a categorical variable or a numerical variable. We wouldn't want to use the exact year numbers, however, since we would want to normalize it in some way. We add both kinds of variables (categorical and normalized numerical) to the dataframe:

```python
# Categorical dummy variables
seasons = pd.get_dummies(df['SEASON_START'])
df = pd.concat([df, seasons], axis=1)
df
```

```python
# Normalize SEASON_START by calculating years since first year in the dataset
df['YEARS_SINCE_2003'] = df['SEASON_START'].apply(lambda x: x - 2003)
df = df.drop(columns=['SEASON_START', 'GAME_DATE_EST'])
df
```

Given that `INFLATION_ADJ_SALARY` is the target variable, we will place it at the end of the dataframe:

```python
cols = df.columns.tolist()
cols.remove('INFLATION_ADJ_SALARY')
cols.append('INFLATION_ADJ_SALARY')
df = df[cols]
df
```

### Standardization

Now that we have all the columns for training, we want to standardize the data. We standardize all numerical columns except for the `PCT` columns, which are already between 0 and 1.

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

ss_columns = ['FGM', 'FGA', 'FG3M', 'FG3A', 'FTM', 'FTA', 'OREB', 'DREB',
              'REB', 'AST', 'STL', 'BLK', 'TO', 'PF', 'PTS', 'PLUS_MINUS']
mm_columns = ['YEARS_SINCE_2003', 'SECS_PLAYED'] # clear lower and upper bound

standard_scaler = StandardScaler()
minmax_scaler = MinMaxScaler()
scaled_df = pd.DataFrame(columns=df.columns)

for col in df.columns:
    if col in ss_columns:
        scaled_df[col] = standard_scaler.fit_transform(df[col].values.reshape(-1, 1)).reshape(-1)
    elif col in mm_columns:
        scaled_df[col] = minmax_scaler.fit_transform(df[col].values.reshape(-1, 1)).reshape(-1)
    else:
        scaled_df[col] = df[col]

scaled_df
```

### Save scaled dataframe

```python
scaled_df.to_csv('../data/preprocessed/stats_salaries_scaled.csv')
```
