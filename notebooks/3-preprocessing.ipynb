## Preprocessing for Model Training

```python
import numpy as np
import pandas as pd

pd.set_option('display.max_columns', None)
```

### Read data and drop/convert columns

```python
df = pd.read_csv("../data/clean/stats_salaries.csv")
df
```

It doesn't look like we need the `PLAYER_NAME`, `TEAM_ABBREVIATION`, or `GAME_DATE_EST` for training, so we can drop that.

```python
df = df.drop(columns=['PLAYER_NAME', 'TEAM_ABBREVIATION', 'GAME_DATE_EST'])
df
```

We also convert minutes played to just seconds:

```python
df['SECS_PLAYED'] = df['MIN'].apply(lambda x: int(x.split(':')[0]) * 60 + int(x.split(':')[1]))
df = df.drop(columns=['MIN'])
df
```

### Convert `START_POSITION` into dummy columns

We convert the `START_POSITION` column into dummy columns first, to indicate what position a player started in. This will also help get rid of all `NaN` values (players who didn't start will have a 0 in all dummy columns).

```python
positions = pd.get_dummies(df['START_POSITION'])
df = pd.concat([df, positions], axis=1).drop(columns=['START_POSITION'])
df
```

### Create season average dataframe

We create a separate season average dataframe to predict only on those values. Since salary is season-based, it might make more sense to predict based on season averages, since each individual game doesn't necessarily reflect the salary a player gets (i.e., a player will have "ups" and "downs").

```python
cols = df.columns.tolist()
cols.remove('PLAYER_ID')
cols.remove('SEASON_START')

aggregations = {col: 'mean' for col in cols if not 'PCT' in col}
# For team name, we don't want an average
aggregations['TEAM_NAME'] = pd.Series.mode
# For plus-minus, sum to know total contribution over course of season
aggregations['PLUS_MINUS'] = 'sum'
avg_df = df.groupby(['PLAYER_ID', 'SEASON_START']).agg(aggregations)

# For percentages, we want to get total made / total attempted
def get_pct_values(s_df):
    fg_pct = s_df['FGM'].sum() / s_df['FGA'].sum() if s_df['FGA'].sum() > 0 else 0
    fg3_pct = s_df['FG3M'].sum() / s_df['FG3A'].sum() if s_df['FG3A'].sum() > 0 else 0
    ft_pct = s_df['FTM'].sum() / s_df['FTA'].sum() if s_df['FTA'].sum() > 0 else 0
    return pd.Series({'FG_PCT': fg_pct, 'FG3_PCT': fg3_pct, 'FT_PCT': ft_pct})
pct_df = df.groupby(['PLAYER_ID', 'SEASON_START']).apply(get_pct_values)

avg_df = avg_df.join(pct_df)
df = df.drop(columns=['PLAYER_ID'])
avg_df = avg_df.reset_index().drop(columns=['PLAYER_ID'])
avg_df
```

### `TEAM_ABBREVIATION` to dummy columns

We also convert the team abbreviation to dummy variables. We expect this could have an effect since different teams have different salary caps.

```python
df_teams = pd.get_dummies(df['TEAM_NAME'])
df = pd.concat([df, df_teams], axis=1).drop(columns=['TEAM_NAME'])
df
```

```python
# If two modes for team name (due to equal # of games played on each team), use the first team
avg_df['TEAM_NAME'] = avg_df['TEAM_NAME'].apply(lambda x: x[0] if isinstance(x, np.ndarray) else x)
avg_df_teams = pd.get_dummies(avg_df['TEAM_NAME'])
avg_df = pd.concat([avg_df, avg_df_teams], axis=1).drop(columns=['TEAM_NAME'])
avg_df
```

### Handle `SEASON_START` data

For `SEASON_START` data, we wouldn't want to use the exact year numbers, since we wouldn't want the first 2000 years to affect the normalization. Instead, we calculate the years passed since the first year in the dataset.

```python
first_season = df['SEASON_START'].min()
first_season
```

```python
# Normalize SEASON_START by calculating years since first year in the dataset
df['YEARS_SINCE_2003'] = df['SEASON_START'].apply(lambda x: x - first_season)
df = df.drop(columns=['SEASON_START'])
df
```

```python
avg_df['YEARS_SINCE_2003'] = avg_df['SEASON_START'].apply(lambda x: x - first_season)
avg_df = avg_df.drop(columns=['SEASON_START'])
avg_df
```

Given that `INFLATION_ADJ_SALARY` is the target variable, we will place it at the end of the dataframe:

```python
cols = df.columns.tolist()
cols.remove('INFLATION_ADJ_SALARY')
cols.append('INFLATION_ADJ_SALARY')
df = df[cols]
avg_df = avg_df[cols]
```

### Save dataframes

```python
df.to_csv('../data/preprocessed/stats_salaries.csv', index=False)
avg_df.to_csv('../data/preprocessed/stats_salaries_avg.csv', index=False)
```
